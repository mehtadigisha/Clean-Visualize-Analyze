```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.shape
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['Amount'] = df['Amount'].astype('int64')
```

```python
df.info()
```

```python
df[['Age','Amount','Orders']].describe()
```

```python
sns.countplot(x='Gender',data=df)
```

```python
ax = sns.countplot(x='Gender',data=df)

for b in ax.containers:
    ax.bar_label(b)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True)
```

```python
sales = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True)
sns.barplot(x='Gender',y='Amount',data=sales)
```

```python
sns.countplot(data=df,x='Age Group',hue='Gender')
```

```python
ax = sns.countplot(data=df,x='Age Group',hue='Gender')

for b in ax.containers:
    ax.bar_label(b)
```

```python
df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
age_ = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=age_)
```

```python
state_ = df.groupby('State',as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(22,6)})
sns.barplot(x='State',y='Orders',data=state_)
```

```python
state_ = df.groupby('State',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(22,6)})
sns.barplot(x='State',y='Amount',data=state_)
```

```python
sns.countplot(x='Product_Category',data=df)
sns.set(rc={'figure.figsize':(25,5)})
```

```python
ax = sns.countplot(x='Product_Category',data=df)
sns.set(rc={'figure.figsize':(20,5)})

for b in ax.containers:
    ax.bar_label(b)
```

```python
df.groupby('Product_Category',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
prod = df.groupby('Product_Category',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.barplot(x='Product_Category',y='Amount',data=prod)
sns.set(rc={'figure.figsize':(20,5)})
```

```python
prod_ = df.groupby('Product_Category',as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.barplot(x='Product_Category',y='Orders',data=prod_)
sns.set(rc={'figure.figsize':(20,5)})
```
